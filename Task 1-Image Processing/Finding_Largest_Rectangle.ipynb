Using python and the openCV library, draw an outline along the edges of the largest rectangle in the image below. The code should determine the largest rectangle.
(Hint: take a look at cv2.findContours, cv2.contourArea, cv2.drawContours)
Rectangles.jpg.jpg


```python
import cv2 #Imports OpenCV-Python which is a library of Python bindings designed to solve computer vision problems.
from google.colab.patches import cv2_imshow #Used for image display in google colab,substitute for cv2.imshow.
```

```python
img=cv2.imread('/content/Rectangles.jpg') #Loading in the image of specific file using imread method of cv2,and the image is Rectangles.jpg here.
o_img=img #variable o_img, to store the original image that will undergo modification throughout the code.
```

```python
gr=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Creating a grayscale version of the image   
```

```python
ed=cv2.Canny(gr,50,200) #Finding the Canny edges, which helps simplify the image for cvs.findContours() function
```

```python
cont, hier= cv2.findContours(ed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) #The cv2.findContours() function finds all external contours in our image, which finds the unique objects in an image.
```

```python
def rec_areas(cont):   #Creating a function that finds all the areas of each of the contours in our image and stores this in a list in the variable, areas. This function returns this list.
   areas= []

   for i in cont:
        area= cv2.contourArea(i) #cv2.contourArea function is used to find size of each object.
        areas.append(area) #Updating areas list with sizes of each of the objects present in the cont list.

   return areas
```

```python
order_cont= sorted(cont, key=cv2.contourArea, reverse= True) #Sorting cont list in descending order of objects-size using the sorted() function. We store this sorted list into the variable, order_cont.Then the list is sorted from largest to smallest by the attribute, reverse=True. To sort a list in ascending order using the attribute, reverse= False, or by specifying nothing in the reverse attribute, as reverse=False, is implied if not specified.
```

```python
largest_rec= order_cont[0] #Now that the list is sorted in descending order, the first item in the list is the largest item, which represents the largest object in our image. We get the largest item with the code, largest_rec= order_cont[0]
```

```python
cv2.drawContours(o_img, largest_rec, -1,(0,255,255),10) #Outlining this largest object in yellow colour with the drawContours() function.
cv2.waitKey(0) #This function will display the window infinitely until any keypress (it is suitable for image display).
cv2_imshow(o_img) #This function is used to display an image in a window.  
```
